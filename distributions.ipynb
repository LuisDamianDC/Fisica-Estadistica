```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
# Binomial distribution for N=5
x = np.random.binomial(5, 0.618, 10000)
plt.hist(x, bins=5)
plt.title('Binomial Distribution for N=5')
plt.show()

# Binomial distribution for N=10
x = np.random.binomial(10, 0.618, 10000)
plt.hist(x, bins=10)
plt.title('Binomial Distribution for N=10')
plt.show()

# Binomial distribution for N=100
x = np.random.binomial(100, 0.618, 10000)
plt.hist(x, bins=100)
plt.title('Binomial Distribution for N=100')
plt.show()
```

```python
# Normal distribution for N=5
mu = 5 * 0.618  # mean = Np
sigma = np.sqrt(5 * 0.618 * (1 - 0.618))  # standard deviation = sqrt(Npq)
x = np.random.normal(mu, sigma, 10000)
plt.hist(x, bins=5)
plt.title('Normal Distribution for N=5')
plt.show()

# Normal distribution for N=10
mu = 10 * 0.618  # mean = Np
sigma = np.sqrt(10 * 0.618 * (1 - 0.618))  # standard deviation = sqrt(Npq)
x = np.random.normal(mu, sigma, 10000)
plt.hist(x, bins=10)
plt.title('Normal Distribution for N=10')
plt.show()

# Normal distribution for N=100
mu = 100 * 0.618  # mean = Np
sigma = np.sqrt(100 * 0.618 * (1 - 0.618))  # standard deviation = sqrt(Npq)
x = np.random.normal(mu, sigma, 10000)
plt.hist(x, bins=100)
plt.title('Normal Distribution for N=100')
plt.show()
```

```python
def factorial(n):
    if n == 0:
        return 1
    else:
        return n * factorial(n-1)
```

```python
N=33
p=0.618
q=1-p
def bin(x):
    return factorial(N)/(factorial(x)*factorial(N-x))*p**x*q**(N-x)

bin_vectorized = np.vectorize(bin)  
x=np.arange(N+1)
plt.title(f'Distribucin Binomial con N={N} y p={p}')
plt.ylabel("Probabilidad")
plt.xlabel("N")
plt.plot(x, bin_vectorized(x))
```

```python
N=5
p=0.618
q=1-p
def y(x):
    return 1 / (np.sqrt(2 * np.pi*N*p*q)) * np.exp(-(x-N*p)**2 / (2*N*p*q))

x=np.linspace(-N*p,3*N*p,100)
plt.title(f'Distribucin Normal con N={N} y p={p}')
plt.ylabel("Probabilidad")
plt.xlabel("N")
plt.plot(x, y(x))
```
